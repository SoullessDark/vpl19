# Тренажер для выучивания стихотворения

Напишите тренажер, который поможет пользователю выучить стихотворение наизусть!
Программа показывает пользователю стихотворение/отрывок, где пропущены случайные слова, а пользователь должен заполнить пропуски.

Например:

    Идет бычок, ___,
    Вздыхает на ходу:
    - Ох, ___ кончается,
    Сейчас я упаду!

В конце программа выдает результат (например, ставит оценку или выводит ободряющие фразы).

Условия:
- Подумайте, как будет удобнее пользователю заполнять пропуски? Сразу показать ему все стихотворение со всеми пропусками или же по частям?
- Введите выбор уровня сложности. Например, на самом простом уровне убираются всего 5 слов из текста, а дальше кол-во пропусков может увеличиваться, а пропускаться могут не только слова, но и сочетания и даже целые строки!
- Определите, какой ответ пользователя вы засчитываете полностью, а какой нет. Будет ли считаться ошибкой одна опечатка, пропуск знака препинания или маленькая буква вместо большой?
- Подумайте об ограничениях для убираемых слов. Например, убирать однобуквенные союзы не очень интересно, так?

```python
poem = '''Gaudeamus igitur,
Juvenes dum sumus!
Post jucundam juventutem,
Post molestam senectutem
Nos habebit humus!
Ubi sunt, qui ante nos
In mundo fuere?
Vadite ad superos,
Transeas ad inferos,
Hos si vis videre!
Vita nostra brevis est,
Brevi finietur.
Venit mors velociter,
Rapit nos atrociter,
Nemini parcetur!
Vivat Academia!
Vivant professores!
Vivat membrum quodlibet!
Vivant membra quaelibet!
Semper sint in flore!
Vivant omnes virgines
Graciles, formosae!
Vivant et mulieres
Tenerae, amabiles,
Bonae, laboriosae!
Vivat et res publica
Et qui illam regunt!
Vivat nostra civitas,
Maecenatum caritas,
Qui nos hic protegunt!
Pereat tristitia,
Pereant dolores!
Pereat Diabolus,
Quivis antiburschius
Atque irrisores!'''
```

```python
import re
from random import sample

def learn(poem):
    level = input('level: ')
    levels = {'check': 1, 'easy': 5, 'medium': 15, 'hard': 25}
    n = levels[level]
    diction = re.sub(r'[^\w]', ' ', poem).split()
    omis = sample(diction, n)
    for elem in omis:
        if poem[poem.index(elem) - 1] in [' ', '\n', '\t'] or poem[poem.index(elem) + 1] in [' ', '\n', '\t']:
            poem = poem.replace(elem, '_'*len(elem))
    print(poem)
    
    while '_' in poem:
        word = input('word ')
        poem = poem[:poem.index('_')] + word + poem[poem.index('_') + len(word):]
    print()
    print('Che brava!')
    print()
    print(poem)

learn(poem)    
```
